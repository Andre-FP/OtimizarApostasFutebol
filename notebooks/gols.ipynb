### Quantidade de escanteios por partida

```python
n_gols = list(df["mandante_Placar"] + df["visitante_Placar"])
```

```python
plt.hist(n_gols, int(max(n_gols)), rwidth=0.8)
plt.xlabel("N√∫mero de gols")
plt.ylabel("Quantidade de partidas")
plt.show()
```

### P(jogo ter == n¬∫ escanteios)

```python
def plot_bar(x, y, title, xlabel, ylabel, arange=(0, 1.05, 0.05)):
    fig, ax = plt.subplots(figsize=(10,6))
    ax.bar(x, y, width=0.8, label="CDF")
    ax.set_xticks(np.arange(len(x) + 1))
    ax.set_yticks(np.arange(arange[0], arange[1], arange[2]))    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
count, bins_count = np.histogram(n_gols, bins=int(max(n_gols)))
pdf = count / sum(count)

title = "P(Jogo ter == n¬∫ escanteios)"
xlabel = "N√∫mero de escanteios"
ylabel = "P(X == n¬∫ de escanteios)"

plot_bar(bins_count[1:], pdf, title, xlabel, ylabel, arange=(0, 0.31, 0.01))
```

### P(jogo ter <= n¬∫ de partidas)

```python
count, bins_count = np.histogram(n_gols, bins=int(max(n_gols)))
pdf = count / sum(count)
cdf = np.cumsum(pdf)
```

```python
title = "P(Jogo ter <= n¬∫ gols)"
xlabel = "N√∫mero de gols"
ylabel = "P(X <= n¬∫ de gols)"
plot_bar(bins_count[1:], cdf, title, xlabel, ylabel)
```

### P(jogo ter <= n¬∫ de partidas)

```python
count, bins_count = np.histogram(n_gols, bins=int(max(n_gols)))
pdf = count / sum(count)

cdf_higher_equal = pdf.copy()
for i in range(len(pdf)):
    for j in range(i + 1, len(pdf)):
        cdf_higher_equal[i] += cdf_higher_equal[j]
```

```python
title = "P(Jogo ter >= n¬∫ gols)"
xlabel = "N√∫mero de gols"
ylabel = "P(X >= n¬∫ de gols)"
plot_bar(bins_count[1:], cdf_higher_equal, title, xlabel, ylabel)
```
